```python
# Load model directly
from transformers import AutoTokenizer, AutoModel

tokenizer = AutoTokenizer.from_pretrained("sentence-transformers/all-MiniLM-L6-v2")
model = AutoModel.from_pretrained("sentence-transformers/all-MiniLM-L6-v2")
```

```python
str1 = "That is a big dog"
str2 = "That is a huge dog"
```

```python
tk1 = tokenizer.encode(str1, return_tensors='pt')
tk2 = tokenizer.encode(str2, return_tensors='pt')

out1 = model(tk1)
out2 = model(tk2)

comp1 = out1.last_hidden_state.mean(1).squeeze()
comp2 = out2.last_hidden_state.mean(1).squeeze()
```

```python
comp1.dot(comp2) / (comp1.norm() * comp2.norm()).abs()
```
